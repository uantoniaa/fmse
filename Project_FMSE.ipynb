# Distributia etichetelor

Relevanță: Arată cât de dezechilibrate sunt clasele (probleme de imbalance în învățarea automată).

```python
labels = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']

label_counts = train_df[labels].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sns.barplot(x=label_counts.index, y=label_counts.values, palette="rocket", hue=label_counts.index)
plt.title("Distribuția etichetelor de toxicitate")
plt.ylabel("Număr comentarii")
plt.show()
```

# Toxic vs Non-Toxic

Relevanță: Oferă o imagine binară a toxic vs non-toxic, ceea ce e util pentru abordări simple de clasificare.

```python
train_df['toxic_comment'] = train_df[labels].sum(axis=1) > 0

colors = ['#A2CFFE', '#FF6961']  
plt.figure(figsize=(6,4))
sns.countplot(x='toxic_comment', hue='toxic_comment', data=train_df, palette=colors, legend=False)
plt.title("Distribuție Toxic vs Non-Toxic")
plt.xticks([0, 1], ['Non-Toxic', 'Toxic'])
plt.ylabel("Număr comentarii")
plt.tight_layout()
plt.show()
```

# Număr de etichete de toxicitate per comentariu

Relevanță: Afișează dacă comentariile sunt marcate cu multiple etichete — clasificare multi-label.

```python

train_df['num_labels'] = train_df[labels].sum(axis=1)

label_dist = train_df['num_labels'].value_counts().sort_index()

colors = sns.color_palette("Set2", n_colors=7)

tick_labels = [
    "Fără etichete (Non-Toxic)",
    "1 etichetă toxică",
    "2 etichete toxice",
    "3 etichete toxice",
    "4 etichete toxice",
    "5 etichete toxice",
    "6 etichete toxice"
]

label_df = label_dist.reset_index()
label_df.columns = ['num_labels', 'count']

plt.figure(figsize=(12,6))
bars = sns.barplot(
    x='num_labels',
    y='count',
    hue='num_labels',
    data=label_df,
    palette=colors,
    legend=False
)

plt.xticks(ticks=range(len(tick_labels)), labels=tick_labels, rotation=45, ha='right')

plt.title("Număr de etichete de toxicitate per comentariu", fontsize=14)
plt.xlabel("Număr etichete", fontsize=12)
plt.ylabel("Număr comentarii", fontsize=12)

for i, bar in enumerate(bars.patches):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height + 200, f'{int(height)}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# Distribuția lungimii comentariilor (în cuvinte)

Relevanță: Observarea variației în lungimea comentariilor — poate fi corelată cu toxicitatea.

```python
train_df['comment_length'] = train_df['comment_text'].apply(lambda x: len(str(x).split()))

counts, bins = np.histogram(train_df['comment_length'], bins=30, range=(0, 500))

colors = sns.color_palette("rocket", n_colors=len(bins)-1)  # una pentru fiecare bin

plt.figure(figsize=(12,5))
for i in range(len(counts)):
    plt.bar(bins[i], counts[i], width=bins[i+1]-bins[i], color=colors[i], align='edge')

plt.title("Distribuția lungimii comentariilor (în cuvinte)", fontsize=14)
plt.xlabel("Număr cuvinte", fontsize=12)
plt.ylabel("Frecvență", fontsize=12)
plt.xlim(0, 500)
plt.tight_layout()
plt.show()
```

# VQC Test

```python
from sklearn.preprocessing import OneHotEncoder

class PatchedOneHotEncoder(OneHotEncoder):
    def __init__(self, **kwargs):
        if 'sparse' in kwargs:
            kwargs['sparse_output'] = kwargs.pop('sparse')
        super().__init__(**kwargs)

import qiskit_machine_learning.algorithms.classifiers.neural_network_classifier as nnc
nnc.OneHotEncoder = PatchedOneHotEncoder
```

```python
from qiskit.circuit.library import TwoLocal, ZZFeatureMap
from qiskit_algorithms.optimizers import COBYLA
from qiskit_algorithms.utils import algorithm_globals

from qiskit_machine_learning.algorithms import VQC
from qiskit_machine_learning.datasets import ad_hoc_data
import pandas as pd
import os
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
X_train_text = train_sample['text'].values
y_train = train_sample['toxic'].values
X_test_text = test_sample['comment_text'].values
y_test = test_sample['toxic'].values

# TF-IDF + PCA
vectorizer = TfidfVectorizer(max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train_text).toarray()
X_test_vec = vectorizer.transform(X_test_text).toarray()

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_vec)
X_test_pca = pca.transform(X_test_vec)

# One-hot encoding etichete
encoder = OneHotEncoder(sparse_output=False)
y_train_oh = encoder.fit_transform(y_train.reshape(-1, 1))
y_test_oh = encoder.transform(y_test.reshape(-1, 1))

# Configurare VQC
algorithm_globals.random_seed = 1376
feature_map = ZZFeatureMap(feature_dimension=2, reps=2, entanglement="linear")
ansatz = TwoLocal(2, ["ry", "rz"], "cz", reps=3)
optimizer = COBYLA(maxiter=100)

vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=optimizer
)
```

```python
%%time
vqc.fit(X_train_pca, y_train_oh)
```

```python
%%time
score = vqc.score(X_test_pca, y_test_oh)

print(f"Testing accuracy: {score:.2f}")
```

# Ad hoc dataset - pentru a incerca VQC pe un dataset mai usor

```python
from qiskit.circuit.library import TwoLocal, ZZFeatureMap
from qiskit_algorithms.optimizers import COBYLA
from qiskit.utils import algorithm_globals

from qiskit_machine_learning.algorithms import VQC
from qiskit_machine_learning.datasets import ad_hoc_data

seed = 1376
algorithm_globals.random_seed = seed

feature_dim = 2 
training_size = 20
test_size = 10

training_features, training_labels, test_features, test_labels = ad_hoc_data(
    training_size=training_size, test_size=test_size, n=feature_dim, gap=0.3
)

feature_map = ZZFeatureMap(feature_dimension=feature_dim, reps=2, entanglement="linear")
ansatz = TwoLocal(feature_map.num_qubits, ["ry", "rz"], "cz", reps=3)
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    optimizer=COBYLA(maxiter=100),
)
vqc.fit(training_features, training_labels)

score = vqc.score(test_features, test_labels)
print(f"Testing accuracy: {score:0.2f}")
```
